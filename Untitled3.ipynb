```python
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv("PhiUSIIL_Phishing_URL_Dataset.csv")
```

```python
X = df['URL'].values  # Select URL column
y = df['label'].values
# urls = df['URL'].tolist()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
def get_sbert_embeddings(texts):
    model = SentenceTransformer('all-MiniLM-L6-v2')
    embeddings = model.encode(texts, batch_size=32, show_progress_bar=True)
    return embeddings
```

```python
embeddings_dict = {}
```

```python
print("Generating SBERT embeddings...")
embeddings_dict['SBERT'] = get_sbert_embeddings(X_train)
embeddings_dict['SBERT'].shape
```

```python
test_embeddings_dict = {}
```

```python
# SBERT
print("Generating SBERT test embeddings...")
test_embeddings_dict['SBERT'] = get_sbert_embeddings(X_test)
test_embeddings_dict['SBERT'].shape
```

```python
X_train_emb = embeddings_dict['SBERT']
X_test_emb = test_embeddings_dict['SBERT']
```

```python
!pip install scikit-learn

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import numpy as np

# Assuming X_train_emb and X_test_emb are already defined as in previous responses

# Determine optimal number of clusters using silhouette score
best_n_clusters = 2  # Initialize with a default value
best_silhouette_score = -1

for n_clusters in range(2, 11):  # Try clusters from 2 to 10
    kmeans = KMeans(n_clusters=n_clusters, random_state=0)
    cluster_labels = kmeans.fit_predict(X_train_emb)
    silhouette = silhouette_score(X_train_emb, cluster_labels)

    if silhouette > best_silhouette_score:
        best_silhouette_score = silhouette
        best_n_clusters = n_clusters

print(f"Optimal number of clusters: {best_n_clusters} (Silhouette score: {best_silhouette_score:.3f})")

# Train KMeans with the optimal number of clusters
kmeans = KMeans(n_clusters=best_n_clusters, random_state=0)
kmeans.fit(X_train_emb)

# Get cluster predictions for training and test data
train_clusters = kmeans.predict(X_train_emb)
test_clusters = kmeans.predict(X_test_emb)

print("KMeans model trained and predictions generated.")
```

```python
import matplotlib.pyplot as plt
cluster_centers = kmeans.cluster_centers_
labels = kmeans.labels_
# Assuming 'labels' contains the cluster assignments for your data points
cluster_0_count = len(labels[labels == 0])
cluster_1_count = len(labels[labels == 1])

# Create a bar chart
plt.bar(['Cluster 0', 'Cluster 1'], [cluster_0_count, cluster_1_count])
plt.xlabel('Cluster')
plt.ylabel('Number of Data Points')
plt.title('Cluster Distribution')
plt.show()
```
